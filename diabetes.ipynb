```python

# alpha argument is used to make the scatter markers transparent.
# By default it values is 1 i.e. no transparency.
plt.scatter(y=df.SkinThickness,x=df.Age,color=['darkorange'],s=800,alpha=0.2)

plt.xlabel("people_Age")
plt.ylabel("people_SkinThickness")
plt.show()
```

```python
plt.scatter(x=df.Age,y=df.BloodPressure,color=['olive'],edgecolor='green',s=800,facecolor='maroon')
plt.ylabel("people_BloodPressure")
plt.xlabel("people_Age")
plt.show()
```

```python
plt.scatter(x=df.BMI,y=df.Age,c=np.random.rand(len(df),3),edgecolor='black',facecolor='maroon',s=500)
plt.xlabel("people_BMI")
plt.ylabel("people_Age")
plt.show()
```

```python
plt.scatter(x=df.Pregnancies,y=df.BMI,c=np.random.rand(len(df),3),s=500)
plt.xlabel("people_Pregnancies ")
plt.ylabel("people_BMI")
plt.show()
```

```python
plt.scatter(x=df.Age,y=df.Pregnancies,c=np.random.rand(len(df),3),facecolor='maroon',s=500)
plt.xlabel("people_Age ")
plt.ylabel("people_Pregnancies ")
plt.show()
```

```python
plt.scatter(x=df.Outcome,y=df.Pregnancies,c=np.random.rand(len(df),3),facecolor='maroon',s=500)
plt.xlabel("people_outcome ")
plt.ylabel("people_Pregnancies ")
plt.show()
```

```python
plt.scatter(x=df.Age,y=df.Pregnancies,c=df.Age,cmap='tab20')

# provide a range for numbers to colors based on the data plotted in the graph
plt.colorbar()

plt.xlabel("people_Age ")
plt.ylabel("people_Pregnancies ")
plt.show()
```

```python
# Ploting the histogram for Glucose  vs Count
plt.hist(df.Glucose,rwidth=0.5)
plt.xlabel("People_Glucose")
plt.ylabel("Age")
```

```python
sns.catplot(y='Pregnancies',x='Age',data=df.sort_values('Pregnancies',ascending=False),kind='boxen',height=6,aspect=3)
plt.show()
```
